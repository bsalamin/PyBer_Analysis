# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_city = pyber_data_df.groupby(['type']).count()['ride_id']
total_rides_city
```

```python
# 2. Get the total drivers for each city type
total_drivers_city = city_data_df.groupby(['type']).sum()['driver_count']
total_drivers_city
```

```python
#  3. Get the total amount of fares for each city type
total_fare_city = pyber_data_df.groupby(['type']).sum()['fare']
total_fare_city
```

```python
#  4. Get the average fare per ride for each city type. 
avgfare_perride_city = pyber_data_df.groupby(['type']).mean()['fare']
avgfare_perride_city
```

```python
# 5. Get the average fare per driver for each city type. 
avgfare_perdriver_city = total_fare_city / total_drivers_city
avgfare_perdriver_city
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame()

pyber_summary_df['Total Rides'] = total_rides_city
pyber_summary_df['Total Drivers'] = total_drivers_city
pyber_summary_df['Total Fares'] = total_fare_city
pyber_summary_df['Average Fare per Ride'] = avgfare_perride_city
pyber_summary_df['Average Fare per Driver'] = avgfare_perdriver_city

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df['Total Rides'] = total_rides_city.map('{:,}'.format)
pyber_summary_df['Total Drivers'] = total_drivers_city.map('{:,}'.format)
pyber_summary_df['Total Fares'] = total_fare_city.map('${:,.2f}'.format)
pyber_summary_df['Average Fare per Ride'] = avgfare_perride_city.map('${:,.2f}'.format)
pyber_summary_df['Average Fare per Driver'] = avgfare_perdriver_city.map('${:,.2f}'.format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
total_fare_day = pyber_data_df.groupby(['type','date']).sum()['fare']
total_fare_day
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
total_fare_day = total_fare_day.reset_index()
total_fare_day
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
total_fare_day_pivot = total_fare_day.pivot(index='date', columns='type', values='fare')
total_fare_day_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
jan_apr_fare_day = total_fare_day_pivot.loc['2019-01-01':'2019-04-29']
jan_apr_fare_day.tail()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
jan_apr_fare_day.index = pd.to_datetime(jan_apr_fare_day.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
jan_apr_fare_day.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
jan_apr_fare_week = jan_apr_fare_day.resample("W").sum()
jan_apr_fare_week.head()
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
import matplotlib.dates as mdates

fig, ax = plt.subplots(figsize=(16, 8))
ax.plot(jan_apr_fare_week)
ax.set_ylabel('Fare ($)',fontsize=14)
ax.set_xticks(pd.date_range(start = "2019-01-01", end = "2019-04-30", freq="MS"))
ax.set_title("Total Fare by City Type")

# Ticks on occurrences of each month:
ax.xaxis.set_major_locator(mdates.MonthLocator())

# Show only month in the x-axis:
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
ax.legend(["Rural","Suburban","Urban"])

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
# Save the figure.
plt.savefig("analysis/PyBer_fare_summary.png")
```
